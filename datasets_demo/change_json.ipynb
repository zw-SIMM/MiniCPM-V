```python
import json
with open("/home/zhangwei/MiniCPM-V/datasets/LLaVA-CC3M-Pretrain-595K/chat.json", 'r') as file:
    data = json.load(file)

data
```

```python

# Function to replace "from" values
def replace_from(data):
    for item in data:
        for conversation in item["conversations"]:
            if conversation["from"] == "human":
                conversation["from"] = "user"
            elif conversation["from"] == "gpt":
                conversation["from"] = "assistant"
            # rename keys
            conversation["role"] = conversation.pop("from")
            conversation["content"] = conversation.pop("value")
    return data

# Replace the "from" values
updated_data = replace_from(data)
```

```python
# Write the updated data to a JSON file
with open('/home/zhangwei/MiniCPM-V/datasets/LLaVA-CC3M-Pretrain-595K/updated_chat.json', 'w') as file:
    json.dump(updated_data, file, indent=4)

print("Data has been updated and written to updated_chat.json")
```
